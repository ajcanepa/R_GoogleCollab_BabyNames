```r
install.packages("babynames")
```

```r
library(babynames)

str(babynames)
```

```r
bruno_df <- babynames[babynames$name == "Bruno", ]
str(bruno_df)
```

```r
Alvaro_df <- babynames[babynames$name == "Alvaro", ]
str(Alvaro_df)
```

```r
plot(x = bruno_df$year,
    y = bruno_df$n,
    type = "l",
    main = "Popularity of the name Bruno")
```

```r
plot(x = Alvaro_df$year,
    y = Alvaro_df$n,
    type = "l",
    main = "Popularity of the name Álvaro")
```
